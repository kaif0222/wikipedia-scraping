```python
# import essential liberies
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

## Geting wilipedia.org table by using requests and beautifulsoup

```python
# adding header, so not to get block


headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    ),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1"
}
```

```python
# getting the wikipedia population page

r = requests.get("https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population_(United_Nations)", headers=headers)
r
```

```python
# extracting html of the page

soup = BeautifulSoup(r.text, "html.parser")
```

```python
# extracting table form it

table = soup.find("table")
```

```python
# extracting the header of table for further use

header = [th.text.strip() for th in table.find_all("th")]
header
```

```python
# getting all rows of the table

rows = []
for tr in table.find_all("tr")[1:]:
  row = [td.text.strip() for td in tr.find_all("td")]
  if row:
    rows.append(row)
```

```python
# converting it into pandas data frame

df = pd.DataFrame(rows, columns=header)
df.head()
```
